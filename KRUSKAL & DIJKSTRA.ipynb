```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices  # Number of vertices in the graph
        self.graph = []  # List to store all edges as [u, v, weight]
        self.heuristics = {}  # Dictionary to store heuristic values for each edg

    def addEdge(self, u, v, w, h=0):  # Function to add an edge along with its heuristic value
        self.graph.append([u, v, w])   # Append edge to graph
        self.heuristics[(u, v)] = h    # Store heuristic value
        self.heuristics[(v, u)] = h  # assuming undirected graph

    def find(self, parent, i):  # Find function for union-find with path compression
        if parent[i] != i:
            parent[i] = self.find(parent, parent[i])  # Recursively find root and compress path
        return parent[i]

    def union(self, parent, rank, x, y):  # Union function using union by rank
        if rank[x] < rank[y]:
            parent[x] = y
        elif rank[x] > rank[y]:
            parent[y] = x
        else:
            parent[y] = x
            rank[x] += 1

    def KruskalMST(self):  # Kruskal's algorithm to find Minimum Spanning Tree (MST)
        result = []  # Store resulting MST edges
        i = 0  # index for sorted edges
        e = 0  # number of edges in result

        # Sort edges based on (weight + heuristic)
        self.graph.sort(key=lambda item: item[2] + self.heuristics.get((item[0], item[1]), 0))

        parent = list(range(self.V))  # Initially, each node is its own paren
        rank = [0] * self.V   # Rank for union by rank

        while e < self.V - 1 and i < len(self.graph):   # Iterate through the sorted edges and build MST
            u, v, w = self.graph[i]
            i += 1
            x = self.find(parent, u)  # Find root of u
            y = self.find(parent, v)  # Find root of v

            if x != y:    # If including this edge doesn't form a cycle
                e += 1
                result.append([u, v, w])  # Include edge in result
                self.union(parent, rank, x, y)    # Union the sets

        minimumCost = sum([weight for _, _, weight in result])
        print("\nEdges in the constructed MST:")
        for u, v, weight in result:
            print(f"{u} -- {v} == {weight}")
        print("Minimum Spanning Tree Cost:", minimumCost)


if __name__ == '__main__':
    print("Name:Prashant Bankar.\nRoll No.:TACO22153")              
    print("Heuristic is used for helping to choose the next edge. "
          "If the heuristic value is low, the edge is preferred.")

    vertices = int(input("Enter the number of vertices: "))
    g = Graph(vertices)
    edges = int(input("Enter the number of edges: "))
    
    print("Enter the edges in the format: u v weight heuristic")
    for _ in range(edges):
        u, v, w, h = map(int, input().split())
        g.addEdge(u, v, w, h)

    g.KruskalMST()
```
